**Data Represented by Line Graph**

Line charts are used to represent the relation between two data on two different axis.

A line Graph is nothing but a way to represent two or more variables in the form of line or curves to visualize the concept and helps to understand it in a better form.

It displays the data that changes continuously concerning time.

In a line graph data points are connected with an edge and data points are represented either with points.

# Simple Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt
xpoints=np.array([0,6])
ypoints=np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
xpoints=np.array([0,4,11,-18,24,36,47,-58,-61])
ypoints=np.array([0,50,15,290,180,-110,270,-210,350])
plt.plot(xpoints,ypoints,color='red',alpha=0.9)
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.show()
```

# Fill the Area Between Two Lines

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(0, 3, 200)
Y1 = X**2 + 3
Y2 = np.sin(X)

plt.plot(X, Y1, '-',lw=4)
plt.plot(X, Y2, '-.',lw=4)

plt.fill_between(X, Y1,Y2, color='red', alpha=.5)

plt.show()
```

About **matplotlib.pyplot.fill_between()**

Refrence Link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_betweenx.html

For use of this function you have to pass one common axis to the both plots,

Otherwise proper area is not printed.

**Syntax:** matplotlib.pyplot.fill_between(common_axis,second_axis_plot1,second_axis_plot2)

**Attributes:**

* color (Default value is blue)

* alpha -> Define opacity of filledcolor

**New Way to plot multiple plot inside the same figure**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4])
y = np.array([[5,6],[10,12],[15,18],[20,24]])

plt.plot(x, y)

plt.fill_between(x,y[:,0],y[:,1],color="red",alpha=0.5)
plt.grid(True)
plt.show()
```

You can not easily fill the graph if you use the above method to draw the plots, i.e. if you pass 2-D array to plot two graph in the same figure.

**Reason** Because if you pass 2-D array to the plot function is normally plot the graphs but during filling time is is confused between the dimensions of y so that it throws an error to avoid this error you have to carefully pass the value of y to the fill_betweeen() function.

if we pass the value as above, in this way we pass the y keeping the dimensions same as x so that it can easily relate x and y both and the proper area.

# Matplotlib Subplot

With the **subplot()** function you can draw multiple plots in one figure:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

About **subplot()**

Refrence Link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

plt.subplot(1, 2, 1)
the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)
the figure has 1 row, 2 columns, and this plot is the second plot.

So, if we want a figure with 2 rows an 1 column (meaning that the two plots will be displayed on top of each other instead of side-by-side), we can write the syntax like this:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

**You can draw as many plots you like on one figure, just descibe the number of rows, columns, and the index of the plot.**

# Title

You can add a title to each plot with the title() function:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

# Super Title

You can add a title to the entire figure with the suptitle() function:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

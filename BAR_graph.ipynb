**Data Representation by BAR graph.**

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.

The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories.

One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

Bar chart helps us to represent the collected data visually.

The collected data can be visualized horizontally or vertically in a bar chart like amounts and frequency.

It can be grouped or single.

It helps us in comparing different items.

By looking at bars, it is easy to say which types of data influence the other.

# Draw a Vertical BAR graph.

**Approach:**

* Importing matplotlib.pyplot as plt

* Creating list x for discrete values on x-axis

* Creating list y  consisting only numeric data for discrete values on y-axis

* Calling plt.bar() function with parameters x,y as plt.bar(x,y)

* Setting x_label() and y_label()

* Setting title() for our bar chart

* Calling plt.show() for visualizing our chart

**Example:-** Number of students enrolled in different courses

courses:- C, C++, python, java

Number of student enrolled:- 43, 23, 34, 56 respectively.

```python
import numpy as np
from matplotlib import pyplot as plt

# Creating a data set
data={"c":43,"c++":23,"python":34,"java":56}

# Convert disctionary keys into a list
courses = list(data.keys())

# Convert dictonary values into a list
values = list(data.values())

#Figure size
fig=plt.figure(figsize=(5,5))

# Plot vertical BAR graph
plt.bar(courses,values,width=0.5,fill=False,hatch='*')
plt.xlabel("Courses Offered")
plt.ylabel("No. of Student Enrolled")
plt.title("Number of student enrolled in different courses")

# Show our plot
plt.show()
```

About **matplot.pyplot.bar()**

Referance Link: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

bar() takes two attribute in sequence

1. values of x-cordinate

2. values of y-cordinate

another optional attributes are:

* **color** (default color is blue)

* **edgecolor** (optional by default any colour is not set)

* **hatch** (used to fill pattern in the bins of bar graph)

* **fill**  (tells the bins is fill or not with the color it has the value 'True' & 'False')

* **width** (width of the bins)
if you provide the value of width = 1 then the gap between the bins become zero or if you increase the value then the bins start overlap to each other
* **align** (default value is center).
   Align the left edges of the bars with the x positions.
   To align the bars on the right edge pass a negative width and align='edge'

**Note:-** In this question a function list() is also used the workin of this function is to create the list only.

# Draw a Horizontal BAR graph.

**Approach:**

* Importing matplotlib.pyplot as plt

* Creating list y   for discrete values on y-axis

* Creating list x  consisting only numeric data for discrete values on x-axis

* Calling plt.barh() function with parameters y,x as plt.barh(y,x)

* Setting x_label() and y_label()

* Setting title() for our bar chart

* Calling plt.show() for visualizing our chart

**Example:-** Number of students enrolled in different courses

courses:- C, C++, python, java

Number of student enrolled:- 43, 23, 34, 56 respectively.

```python
import numpy as np
from matplotlib import pyplot as plt

# Creating a data set
data={"c":43,"c++":23,"python":34,"java":56}

# Convert disctionary keys into a list
courses = list(data.keys())

# Convert dictonary values into a list
values = list(data.values())

# Figure size
fig=plt.figure(figsize=(5,10))

# plot Horizontal BAR garph
plt.barh(courses,values,color="yellow",align="center",hatch='///')
plt.ylabel("Courses Offered")
plt.xlabel("No. of Student Enrolled")
plt.title("Number of student enrolled in different courses")

# Show our plot
plt.show()
```

# Multiple Bar Graphs

```python
# set width of bar
barwidth = 0.25
fig = plt.subplots(figsize=(12,8))

# set height of bar
IT = [12,30,1,8,22]
ECE = [28,6,16,5,10]
CSE = [29,3,24,25,17]

# set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barwidth for x in br1]
br3 = [x + barwidth for x in br2]

# Make the plot
plt.bar(br1,IT,color='red',width=barwidth,edgecolor='black',label='IT')
plt.bar(br2,ECE,color='green',width=barwidth,edgecolor='black',label='ECE')
plt.bar(br3,CSE,color='blue',width=barwidth,edgecolor='black',label='CSE')

# adding Xticks
plt.title('Stacked BAR plot')
plt.xlabel('Branch',fontweight='bold',fontsize=15)
plt.ylabel('student passed',fontweight='bold',fontsize=15)
plt.xticks([r+barwidth for r in range(len(IT))],['2015','2016','2017','2018','2019'])
plt.legend()
plt.show()
```

# Stacked BAR Graph

Stacked bar plots represent different groups on top of one another.

The height of the bar depends on the resulting height of the combination of the results of the groups.

It goes from the bottom to the value instead of going from zero to value.

**A traditional bar chart displays a single categorical variable in the form of bars.
A stacked bar chart shows two categorical variables.**

**Approach:**

1. Import Library (Matplotlib)

2. Import / create data.

3. Plot the bars in the stack manner.

Simple Example

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

Another Example

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

**Example:-** The following bar plot represents the contribution of boys and girls in the team.

```python
N = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,bottom = boys, yerr = girlStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```
